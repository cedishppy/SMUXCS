```python

prefecs = pd.read_csv("../data/japan_prefectures_basemap.csv")
df_right = pd.read_csv("../data/Japan_Data.csv")
```

```python
from pandas import DataFrame
correct_df = df_right.iloc[:, 2:4]
correct_df = correct_df.values.tolist()
correct_df
```

```python
prefecs = prefecs.loc[:,'name_1']
prefecs
```

```python
checking_list = []
for i in prefecs:
    checking_list.append([i,0])
new_prefecs = checking_list
```

```python
new_prefecs
```

```python
for rows in correct_df:
    prefecture = rows[1]
    value = rows[0]
    for prefectures in new_prefecs:
        counter = 0
        if prefecture == "Hyogo":
            if prefectures[0] == 'Hygo':
                prefectures[1] += value
            counter +=1
        if prefecture == "Nagasaki":
            if prefectures[0] == 'Naoasaki':
                prefectures[1] += value
            counter +=1
        elif prefecture == prefectures[0]:
            prefectures[1] += value
            counter +=1
    
```
