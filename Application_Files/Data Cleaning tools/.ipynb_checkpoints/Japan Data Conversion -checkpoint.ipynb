```python
df = pd.read_csv("../data/Japan_Data.csv")
prefecs = pd.read_csv("../data/japan_prefectures_basemap.csv")
df_right = pd.read_csv("../data/Japan_Data_right_table.csv")
```

```python
value_pref = df.iloc[:, 2:3]
```

```python
old_d = df.iloc[:, 3:4]
new_d = df.iloc[:, 3:4]
```

```python
count_row_o = old_d.shape[0] 
print(count_row_o)

count_row_n = new_d.shape[0] 
print(count_row_n)

count_row_v = value_pref.shape[0] 
print(count_row_v)
```

```python
# Removal of inaccurate values 
listtodf = []
for i in range (5484):
    o = old_d.iloc[i,0]
    n = new_d.iloc[i,0]
    val = value_pref.iloc[i,0]
    if o == "NA":
        correct = n
    else:
        correct = o
    listtodf.append([val,correct])
```

```python
from pandas import DataFrame
correct_df = DataFrame(listtodf,columns=['Value','Prefrecture'])
```

```python
correct_df = correct_df.values.tolist()
```

```python
prefecs = prefecs.loc[:,'name_1']
prefecs
```

```python
checking_list = []
for i in prefecs:
    checking_list.append([i,0])
new_prefecs = checking_list
```

```python
new_prefecs
```

```python
for rows in correct_df:
    prefecture = rows[1]
    value = rows[0]
    for prefectures in new_prefecs:
        counter = 0
        if prefecture == "Hyogo":
            if prefectures[0] == 'Hyōgo':
                prefectures[1] += value
            counter +=1
        elif prefecture == prefectures[0]:
            prefectures[1] += value
            counter +=1
    
```
